```python
import pandas as pd
```

```python
dataset_path = 'datasetCombined.csv'

column_names = ['Cografya', 'Yerlesim', 'AracYogunlugu', 'YayaYogunlugu', 
                'VeriHizi', 'Guvenilirlik', 'Gecikme', 'Kapsama', 'Standart']

dataset = pd.read_csv(dataset_path, header=None)
```

```python
duplicate = dataset[dataset.duplicated()]
print('Number of duplicated rows:', len(duplicate))
```

```python
rows, cols = dataset.shape
label_idx = cols - 1

for i in range(rows):
    if dataset.iat[i, label_idx] == 2:
        dataset.iat[i, label_idx] = 0
```
